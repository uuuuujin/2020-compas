### 2.2. 공간 정보의 시각화

배경 맵은 국토교통부에서 제공하는 vworld 를 기준으로 표현합니다.

```python
## 배경맵을 시각화합니다.
import folium

center = (37.658370, 126.831949); zoom = 15

m = folium.Map(location=center,
           zoom_start=zoom,
           tiles='http://api.vworld.kr/req/wmts/1.0.0/YOUR_API_KEY/Base/{z}/{y}/{x}.png',
           attr='고양시'
)

m
```

다음은 제공된 데이터를 vworld 에 표현한 예입니다.

```python
## 고양시 체육시설 현황 정보를 살펴보겠습니다.
import pandas as pd
df = pd.read_csv('./data/24.고양시 체육시설 현황 정보.csv')
df.head()
```

```python
## 해당 시설들의 위치를 지도에서 확인합니다.
for index, row in df.iterrows():
    folium.Marker([row['Y'], row['X']], 
                  popup=row['FAC_NAME'],
                  icon=folium.Icon(icon='blue')).add_to(m)
m
```
