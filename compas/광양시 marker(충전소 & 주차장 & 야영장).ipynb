```python
#충전소 : ChargingStation

chargingStation = pd.read_csv(p.joinpath('01.광양시_충전기설치현황.csv'))
chargingStation
```

```python
#주차장 : parkingLot

parkingLot = pd.read_csv(p.joinpath('02.광양시_주차장_공간정보.csv'))
parkingLot
```

```python
#주차장 : parkingLot
campsite = pd.read_csv(p.joinpath('05.광양시_대중집합시설_야영장.csv'))
campsite
```

```python
lat = chargingStation['lat'].mean()
lon = chargingStation['lon'].mean()
lat, lon
```

```python
m = folium.Map([lat,lon], zoom_start=11)

for i in chargingStation.index:
    cs_lat = chargingStation.loc[i,"lat"]
    cs_lon = chargingStation.loc[i,'lon']
    title = chargingStation.loc[i,"충전소명"] + " - " + chargingStation.loc[i,"충전소위치"]
    
    folium.Marker([cs_lat, cs_lon],icon=folium.Icon(color='blue'), tooltip=title).add_to(m)
m

for i in parkingLot.index:
    pl_lat = parkingLot.loc[i,"lat"]
    pl_lon = parkingLot.loc[i,'lon']
    title = parkingLot.loc[i,"주차장명칭"] + " - " + parkingLot.loc[i,"주소"]
    
    folium.Marker([pl_lat, pl_lon], icon=folium.Icon(color='darkpurple'), tooltip=title).add_to(m)
m

for i in campsite.index:
    cp_lat = campsite.loc[i,"lat"]
    cp_lon = campsite.loc[i,'lon']
    title = campsite.loc[i,"명칭"] + " - " + campsite.loc[i,"주소"]
    
    folium.Marker([cp_lat, cp_lon], icon=folium.Icon(color='lightgreen'), tooltip=title).add_to(m)
m
```
